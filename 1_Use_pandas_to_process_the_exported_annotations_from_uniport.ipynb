# Analysis of the AlphaFold predictions of four-domain MetH
## 1. Use pandas to process the exported annotations from uniport


```python
import pandas as pd
import pickle
```

```python
# import the uniprot ID mapping results
df = pd.read_csv (r'MetH_4d_uniprot.csv')
display(df)
```

```python
# filter out the items that have no domain or binding site information
df1 = df[df['Domain'].notnull()]
df1 = df1[df1['Binding_site'].notnull()]
print(len(df)-len(df1),'items were filtered out')
display(df1)
```

```python
# split the domain column at ; and filter out the annotation of different domains
df2 = df1.Domain.str.rsplit(";", expand=True)
df2 = df2[df2[1].str.contains('Hcy')]
df2 = df2[df2[4].str.contains('Pterin')]
df2 = df2[df2[7].str.contains('B12-binding N-terminal')]
df2 = df2[df2[10].str.contains('B12-binding')]
df2 = df2[df2[13].str.contains('AdoMet')]
print(len(df1)-len(df2),'items were filtered out')
df5 = [];
for i in [0,3,6,9,12]:
    df3 = []
    df4 = []
    df3 = df2[i].str.split(expand=True)
    df4 = df3[1].str.rsplit("..", n=1, expand=True)
    df5.append(df4) 
result1 = pd.concat(df5,axis=1)
result1.columns = ['Hcy_1', 'Hcy_2', 'MTHF_1', 'MTHF_2', 'cap_1', 'cap_2', 'B12_1', 'B12_2', 'AdoMet_1', 'AdoMet_2']
display(result1)
```

```python
# Recombine the dataframes so we can filter out the entries that do not have the correct domain information (This is a weird way of doing it, there is a better way, but since I wrote this first and it worked so I do not want to change)
result2 = pd.concat([df1.Entry,result1,df1.Binding_site],axis=1)
Domain_clean = result2.copy()
for i in range(1,11):
    Domain_clean = Domain_clean[Domain_clean.iloc[:,i].notnull()]
    Domain_clean = Domain_clean[Domain_clean.iloc[:,i].str.isnumeric()]
display(Domain_clean)
```

```python
# split the Binding_site column at ;
df2b = Domain_clean.Binding_site.str.rsplit(";", expand=True)
display(df2b)
```

```python
# Define the function to find the position of the binding sites
def find_binding_sites_positions(pandalist,binding_sites_key_word):
    position = []
    row_n = []
    m = 0
    for i in range(0,len(pandalist)):
        position_i = [i]
        columns_p = []
        mask = pandalist.iloc[i,:].str.contains(binding_sites_key_word,na=False) 
        columns_p = pandalist.columns[mask].to_numpy()
        position_i.append(columns_p)
        position.append(position_i)
        if not columns_p.size:
            index_p = pandalist.iloc[[i]].index.to_numpy()
            print(f'No such information in row {i}, the index of the row is {index_p}')
            row_n.append(i)
            m +=1
        #else:
            #print(f'Information found in row {i}')
                
    print(m,'entries not available')
    return position, row_n
    
```

```python
# Find the column positions of cobalamin binding sites
cob_p,cob_n = find_binding_sites_positions(df2b,f'methylcob')
print(len(cob_p),'items were accessed in total')
```

```python
# Find the column positions of SAM binding sites
sam_p,sam_n = find_binding_sites_positions(df2b,f'S-adenosyl-L-methionine')
print(len(sam_p),'items were accessed in total')
```

```python
# Find the column positions of Hcy binding sites
Hcy_p,Hcy_n = find_binding_sites_positions(df2b,f'Zn')
print(len(Hcy_p),'items were accessed in total')
print(Hcy_n,'are the positions of the unavilable rows')
```

```python
# Define the function of getting the binding site positions
def get_binding (panda,position,panda_domain_start,panda_domain_end):
    df5b = []
    p_n = []
    n_n = []
    w_n = []
    no_inside_binding_site = []
    for i in range(0,len(panda)):
        m1 = 0
        m2 = 0
        m3 = 0
        df3b = []
        binding_number = []
        position_b = position[i][1]-1
        df3b = panda.iloc[i,position_b]
        d_start = int(panda_domain_start.iloc[i])
        print(d_start)
        
        d_end = int(panda_domain_end.iloc[i])
        print(d_end)
        for j in range(0,len(df3b)):
            try:
                # why to use rsplit() in addition to split(): some binding sites are a range of amino acids instead of not a single one, so it will be like 770..774
                binding_number_s = df3b.iloc[j].split()[1].rsplit('..')
                if len(binding_number_s)>1:
                    print(i,'more than one binding sites')
                    range_n = int(binding_number_s[-1])-int(binding_number_s[0])
                    for k in range(0,range_n):
                        if int(binding_number_s[0])+k>=d_start and int(binding_number_s[0])+k<=d_end:
                            binding_number.append(int(binding_number_s[0])+k)
                        if int(binding_number_s[0])+k<d_start:
                            print(i,'the binding site is in the previous domain')
                            m1 = 1
                        if int(binding_number_s[0])+k>d_end:
                            print(i,'the binding site is in the next domain')
                            m2 = 1
                if len(binding_number_s)==1:
                    if int(binding_number_s[0])>=d_start and int(binding_number_s[0])<=d_end:
                        binding_number.append(int(binding_number_s[0]))
                    if int(binding_number_s[0])<d_start:
                        print(i,'the binding site is in the previous domain')
                        m1 = 1
                    if int(binding_number_s[0])>d_end:
                        print(i,'the binding site is in the next domain')
                        m1 = 1
                if len(binding_number_s)==0:
                    print(i,'wrong')
                    m3 = 1
            except:
                print(i)
        if m1:
            p_n.append(i)
        if m2:
            n_n.append(i)
        if m3:
            w_n.append(i)
        if not len(binding_number):
            no_inside_binding_site.append(i)
            
        df5b.append(binding_number)
    print(len(p_n),len(n_n),len(w_n))
    return df5b, no_inside_binding_site

# Delete the element that does not have Hcy binding information
def remove_no_hcy (ps, hcy_n):
    for i in range(0,len(hcy_n)):
        ps.pop(hcy_n[i]-i)
```

```python
# Get the position of Zinc binding sites
Zinc_ps,no_i_zinc = get_binding(df2b,Hcy_p,Domain_clean.Hcy_1,Domain_clean.Hcy_2)
print(no_i_zinc)
print(len(Zinc_ps))
print(Zinc_ps[Hcy_n[0]])
remove_no_hcy(Zinc_ps,Hcy_n)
print(len(Zinc_ps))
print(Zinc_ps[Hcy_n[0]])
```

```python
# Get the position of B12 binding sites
B12_ps,no_i_B12 = get_binding(df2b,cob_p,Domain_clean.B12_1,Domain_clean.B12_2)
print(no_i_B12)
print(len(B12_ps))
remove_no_hcy(B12_ps,Hcy_n)
print(len(B12_ps))
```

```python
# Get the position of SAM binding sites
SAM_ps,no_i_sam = get_binding(df2b,sam_p,Domain_clean.AdoMet_1,Domain_clean.AdoMet_2)
print(no_i_sam)
print(len(SAM_ps))
remove_no_hcy(SAM_ps,Hcy_n)
print(len(SAM_ps))
```

```python
# Define the function to convert the position of binding sites to string
def position_to_string (binding_sites,array_name):
    binding_site_string = []
    no_binding = []
    for i in range(0,len(binding_sites)):
        binding_site_string_s = []
        if not len(binding_sites[i]):
            no_binding.append(i)
            print('Number',i,'item in',array_name, 'has no such binding site')
        for j in range(0,len(binding_sites[i])):
            binding_site_string_s.append(str(binding_sites[i][j]))
        binding_site_string.append('+'.join(binding_site_string_s))
    return binding_site_string,no_binding
                
            
```

```python
# Convert the binding sites information to string
B12_ps_string,no_b12 = position_to_string (B12_ps,'B12_ps')
SAM_ps_string,no_sam = position_to_string (SAM_ps,'SAM_ps')
Zinc_ps_string,no_zinc = position_to_string (Zinc_ps,'Zinc_ps')
```

```python
# Delete the positions with no sam binding site
remove_no_hcy(B12_ps_string,no_sam)
remove_no_hcy(SAM_ps_string,no_sam)
remove_no_hcy(Zinc_ps_string,no_sam)
# store the virables to pkl files
B12_f = open('B12_ps_string', 'wb')
pickle.dump(B12_ps_string, B12_f) # serialize the list
B12_f.close()

SAM_f = open('SAM_ps_string', 'wb')
pickle.dump(SAM_ps_string, SAM_f) # serialize the list
SAM_f.close()

Zinc_f = open('Zinc_ps_string', 'wb')
pickle.dump(Zinc_ps_string, Zinc_f) # serialize the list
Zinc_f.close()
```

```python
# Delete the rows that do not have full binding site information
df2b_new0 = df2b.drop(df2b.index[Hcy_n])
df2b_new = df2b_new0.drop(df2b_new0.index[no_sam])
display(df2b_new)
```

```python
# Recombine the whole filtered dataframe
MetH_4d_info = pd.concat([df1.Entry[df2b_new.index],result1.loc[df2b_new.index]],axis=1)
MetH_4d_info.columns = ['Uniprot_ID', 'Hcy_1', 'Hcy_2', 'MTHF_1', 'MTHF_2', 'cap_1', 'cap_2', 'B12_1', 'B12_2', 'AdoMet_1', 'AdoMet_2']
MetH_4d_info.to_pickle("MetH_4d_info.pkl")
display(MetH_4d_info)
```
