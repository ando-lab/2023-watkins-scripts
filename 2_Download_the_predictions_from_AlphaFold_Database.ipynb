# Analysis of the AlphaFold predictions of four-domain MetH
## 2. Download the predictions from AlphaFold Database https://alphafold.ebi.ac.uk
### The pdbs used is provided as a .zip file, since downlading all the files takes a while. You can unzip the file, and all the pdbs should be there. The pdb files provided contain more pdbs than I used. If you want to test, you can delete some of the files, the code will download the missing pdb files for you.


```python
import subprocess
import requests
import os
import subprocess
import pandas as pd
```

```python
# load the variables
MetH_4d_info = pd.read_pickle("MetH_4d_info.pkl")
display(MetH_4d_info)
```

```python
# download the structures from the AlphaFold database
Folder_name = 'MetH_Alphafold_database_4d' # Note: Check the folder name and make sure it matches what you have. The folder name cannot have any space
if not os.path.exists(f'{Folder_name}'):
    os.mkdir(f'{Folder_name}')
for i in range(0,len(MetH_4d_info.Uniprot_ID)):
    uniprot_ID = MetH_4d_info.Uniprot_ID.iloc[i]
    database_version = 'v3'
    model_url = f'https://alphafold.ebi.ac.uk/files/AF-{uniprot_ID}-F1-model_{database_version}.pdb'
        
    if not os.path.exists(f'{Folder_name}/{uniprot_ID}.pdb'):
        try:
            r = requests.get(model_url)
            r.raise_for_status()
            r.encoding = r.apparent_encoding
            os.system(f'curl {model_url} -o {uniprot_ID}.pdb')
            subprocess.run([f"mv {uniprot_ID}.pdb {Folder_name}"], shell=True)
            print(f'{uniprot_ID}.pdb was downloaded successfully')
        except:
            print('Internet connection error or no prediction available in Alphafold database for',uniprot_ID)
```
